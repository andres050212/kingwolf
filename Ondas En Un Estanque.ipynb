```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin
salida2=open("Ondas_Est.dat3", "w")
pixeles = 1000
tamaox = 100. # cm 
lda = 5.0 # cm
lda1 = 3.0
k = 2.*pi / lda
k1 = 2.*pi / lda1
salto = tamaox / pixeles # cm/pixel
zeta0 = 1.0 # cm
x0 = 25. # cm
y0 = 50. # cm
x1 = 75.
y1 = 50.
for i in range(0,pixeles): 
    for j in range(0,pixeles): 
        r=sqrt((x0-i*salto)**2 + (y0-j*salto)**2)
        e=zeta0*sin(k*r)
        r1=sqrt((x1-i*salto)**2 + (y1-j*salto)**2)
        e1=zeta0*sin(k1*r1)
        e21=e+e1
        salida2.write("%0.2f "% (e+e1))
        
    salida2.write("\n")
                                          
salida2.close()

from pylab import imshow, show, xlabel, ylabel,cool
from numpy import loadtxt

data = loadtxt("Ondas_Est.dat3", float)
print(data.shape)
imshow(data)
imshow(data, origin="lower", extent=[0, 100, 0, 100], aspect=1.0)
cool()
xlabel("Eje X cm")
ylabel("Eje Y cm")
show()
```
