```python
from pylab import plot, imshow, show, xlabel, ylabel
from numpy import loadtxt, array, zeros, sqrt, pi, sin
import matplotlib.pyplot as plot

UA = 1.495e11 #m
GM = 4*pi**2
m = 10e15


dt = 0.02
vel = []
rad = []
time = []
Energy = []
rn = 1
vn = 2*pi
theta = []
the = 0



for i in range(201):
    rn = rn+(vn*dt)
    a = GM/(rn**2)
    vn = vn+(a*dt)
    Et = -(GM*m)/(2*rn)
    the = the+(vn/rn)*dt
    rad.append (rn)
    time.append (dt*i)
    Energy.append (Et)
    theta.append (the)
    
plot.plot (time,rad)
show ()
plot.plot (rad, Energy)
show ()
plot.polar(theta,rad,'+')
plot.show()
    
```
